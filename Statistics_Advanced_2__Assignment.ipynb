## Question 1: What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical procedure used to make a decision about a population based on sample data. It involves creating a "null hypothesis" (the default assumption) and an "alternative hypothesis" (the claim being tested) and then using sample data to determine whether there is enough evidence to reject the null hypothesis. This process is crucial for deciding if an observed effect in a sample is a real population-level effect or just due to random chance.

## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

- The null hypothesis (H0) is a statement of no effect or no relationship between variables, assuming any observed difference is due to chance, while the alternative hypothesis (H1) is a statement that claims a significant effect or relationship exists. Essentially, researchers test the null hypothesis to find evidence that proves it false, which in turn supports the alternative hypothesis.

## Null hypothesis (H0)     

- Definition: A default or "no effect" statement, meaning there is no relationship or difference between the variables being studied.

- Goal of testing: To be disproven or rejected.

- Mathematical symbol: Typically includes an equality sign (=,<=, >=).

- Example: A new drug has no effect on blood pressure, or there is no difference in the average height of two groups of students.

## Alternative hypothesis (H1)     

- Definition: A statement that there is a significant relationship or effect between the variables. It is what the researcher is trying to prove.

- Goal of testing: To be supported if the null hypothesis is rejected.

- Mathematical symbol: Uses an inequality sign (!=, <, >).

- Example: The new drug does have an effect on blood pressure, or the average height of the two groups of students is statistically different.

## Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

 - The significance level, or alpha (alpha), is a threshold for deciding whether to reject the null hypothesis. It represents the probability of a Type I error—incorrectly rejecting a true null hypothesis. The outcome of a hypothesis test is determined by comparing the test's p-value to this pre-set alpha. If the p-value is less than alpha, the null hypothesis is rejected; otherwise, it is not rejected.

## Role in deciding the outcome     
     
- Setting the threshold: The significance level is a value chosen by the researcher before the test begins, often set at common levels like 0.05 (5%), 0.01 (1%), or 0.10 (10%). This value acts as the benchmark for statistical significance.

- Comparing the p-value: After the data is analyzed, a p-value is calculated. This p-value is the probability of observing data as extreme as, or more extreme than, the sample results, assuming the null hypothesis is true.

- Making the decision: The p-value is compared to the significance level (alpha):

    - If (p < alpha): The result is considered statistically significant. The evidence is strong enough to reject the null hypothesis in favor of the alternative hypothesis. This means the observed data is unlikely to have occurred by chance alone.
    
    - If (p >= alpha): The result is not statistically significant. There is not enough evidence to reject the null hypothesis. The data could have occurred by chance.
    
 ## Example

- Scenario: A researcher tests a new drug and sets the significance level at (alpha =0.05).

- Result: The analysis yields a p-value of (0.02).

- Outcome: Since (0.02 < 0.05), the researcher rejects the null hypothesis. The results are statistically significant at the 5% level, suggesting the drug has a real effect.

## Question 4: What are Type I and Type II errors? Give examples of each.
##**Type I error:**
-  Type I Error (False Positive) Happens when we reject H₀ even though it is true.

- In simple words: we think there is an effect/difference, but actually, there isn’t.

- Probability of this error = α (significance level).

Example:
- Court case: An innocent person (H₀ true) is declared guilty → Type I error.

 - Medicine trial: We conclude a new drug works better, but in reality, it doesn’t.

##**Type II error:**

- Type II Error (False Negative) Happens when we fail to reject H₀ even though H₁ is true.

- In simple words: we miss a real effect/difference.

- Probability of this error = β (related to test’s power z)= 1 − β.

Example:

- Court case: A guilty person (H₁ true) is declared innocent → Type II error.

- Medicine trial: We conclude the new drug has no effect, but in reality, it works.

## Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

|       Feature                    | Z-Test                  |   T-Test           |
 ----------------------------       | -------------------        | ----------------
Population Standard Deviation    |   Known        | Unknown
Sample Size                      | Large (n>=30)  | Small (n<=30)
Underlying Distribution          | Standard Normal (z) distribution | 	T-distribution
 Primary Use Case | Comparing a sample mean to a population mean when population standard deviation is known and sample size is large | Comparing the means of two groups when population standard deviation is unknown and sample size is small

# When to use a T-test:

- Use when: The population standard deviation is unknown, and your sample size is small (typically n<30).

- Why: A T-test accounts for the extra uncertainty that comes with estimating the population standard deviation from a small sample. The T-distribution is more spread out than the standard normal distribution, which is necessary for small samples.

- Example: A teacher compares the reading scores of 28 students who played a reading game to 28 who didn't. Since the sample size is small (n=28), a T-test is appropriate.

# When to use a Z-test:     

- Use when: The population standard deviation is known, and your sample size is large (typically n >= 30).

- Why: With a large sample, you can reliably estimate the population standard deviation, and the standard normal distribution (Z-distribution) is a suitable test statistic.

- Example: A quality control manager is comparing the weight of a large batch of products (n=100) against a known historical population average and standard deviation.








## Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 10000  # number of samples

# Generate binomial distribution samples
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

## Question 7: Implement hypothesis testing using Z-statistics for a sample dataset inPython. Show the Python code and interpret the results.

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)   # sample std (unbiased)
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Sample Size:", n)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

## Question 8: Write a Python script to simulate data from a normal distribution andcalculate the 95% confidence interval for its mean. Plot the data using Matplotlib.




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # true mean and std deviation
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha/2)  # Z critical value

margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 3: Plot the data
plt.figure(figsize=(8,5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, color='skyblue')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")

plt.title("Histogram of Simulated Normal Data with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## Question 9: Write a Python function to calculate the Z-scores from a dataset andvisualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.



```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.
    Z = (x - mean) / std
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z_scores = (data - mean) / std
    return z_scores, mean, std

# Example dataset
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)  # N(50, 5^2)

# Calculate Z-scores
z_scores, mean, std = calculate_z_scores(data)

print("Original Mean:", mean)
print("Original Std Dev:", std)
print("First 5 Z-scores:", z_scores[:5])

# Plot histogram of standardized data
plt.figure(figsize=(8,5))
plt.hist(z_scores, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (0 after standardization)")
plt.title("Histogram of Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
