###1. What is unsupervised learning in the context of machine learning ?

-  **Unsupervised learning** is a type of machine learning that finds hidden patterns, structures, and relationships in unlabeled data, without predefined outputs or human guidance, focusing on discovering insights like data groupings (clustering) or simplifying data (dimensionality reduction). Unlike supervised learning, it explores raw data to infer its own rules, ideal for tasks where labeling data is costly, such as customer segmentation, anomaly detection, or market basket analysis.


###2. How does K-Means clustering algorithm work ?

- **K-means clustering** works by iteratively partitioning data into k groups (clusters) by assigning each data point to the nearest cluster centroid (center) and then recalculating those centroids as the mean of their assigned points, repeating until cluster assignments stabilize, aiming to minimize the distance between points and their cluster centers. It's an unsupervised method for finding natural groupings in unlabeled data, requiring the user to define the number of clusters, k.  

###Here's a step-by-step breakdown:
1. Choose k : Decide on the number of clusters (k) you want to find in your data.

2. Initialize Centroids: Randomly place k points (centroids) in the data space, or select k initial data points as centroids.

3. Assign Points (E-Step): Calculate the distance (usually Euclidean) from each data point to every centroid and assign each point to the cluster with the nearest centroid.

4. Update Centroids (M-Step): Recalculate the position of each centroid as the average (mean) of all data points now assigned to that cluster.

5. Repeat: Go back to Step 3 and repeat the assignment and update steps. The algorithm converges (stops) when the centroids stop moving significantly, meaning points are consistently assigned to the same clusters.

The goal is to find cluster centers that best represent the data, minimizing the "inertia" or within-cluster sum of squared distances, revealing natural groupings based on feature similarity.

###3. Explain the concept of a dendrogram in hierarchical clustering ?

- A **dendrogram** is a tree-like diagram visualizing the hierarchical relationships formed by hierarchical clustering, showing how data points or clusters merge (agglomerative) or split (divisive) at different similarity/distance levels, with branches indicating closeness (shorter branches = more similar) and the y-axis representing the distance/dissimilarity at which merges occur, allowing you to cut the tree to define the number of clusters.


###Key Components & Interpretation:

- Tree Structure: Starts with individual data points (leaves) and connects them into clusters, forming branches (clades).

- Y-Axis (Height): Represents the distance or dissimilarity between merged clusters; taller lines mean greater distance/less similarity.

- X-Axis: Lists the individual data points or clusters.

- Merging/Splitting: In agglomerative (bottom-up) clustering, it shows when clusters join; in divisive (top-down), when they split.

- Cluster Formation: Points joined by short branches are very similar; points joined by tall branches are dissimilar.


###How to Use It:

1. Identify Clusters: Draw a horizontal line across the dendrogram.

2. Count Clusters: The number of vertical lines crossed by your horizontal line determines the number of clusters.

3. Define Clusters: All points below the line form a cluster.

###Example

- If a line cuts through two short, closely merged branches, you get two tight clusters. If it cuts much higher, crossing a single long branch, you might get one large cluster.

###6. When would you use Silhouette Score in clustering ?

###Key Scenarios for Using Silhouette Score:

- **Evaluating Cluster Quality:** It provides a quantitative measure (from -1 to +1) of cohesion (points close in their cluster) and separation (clusters far apart), giving a clear signal of cluster "goodness".

- **Comparing Clustering Algorithms:** Use it to compare results from different algorithms (e.g., K-Means vs. DBSCAN) or different parameter settings for the same algorithm.

- **Identifying Misclassified Points:** Negative silhouette scores for individual points highlight data points that are likely in the wrong cluster, indicating potential outliers or poorly formed groups.

- V**isualizing Cluster Structure:** Silhouette plots help visualize the density and separation of each cluster, revealing if some clusters are much larger or less cohesive than others, notes scikit-learn.

###When It's Especially Useful:

- When clusters aren't perfectly distinct and you have boundary cases.

- To add rigor and objectivity to unsupervised learning, where traditional accuracy metrics don't apply.

###Limitations:

- It works best for convex-shaped clusters (like those K-Means finds) and might struggle with irregularly shaped or varying-sized clusters.

###7. What are the limitations of Hierarchical Clustering?

- **Hierarchical clustering's** main limitations are its high computational cost (scalability issues), making it slow for large datasets; irreversible, greedy decisions that can lead to suboptimal clusters; sensitivity to noise and outliers, which distort the hierarchy; and difficulty in interpreting large dendrograms, especially with many clusters. It struggles with huge datasets and relies heavily on chosen distance/linkage methods, requiring user judgment for interpretation.

### Key Limitations:

- **Scalability/Computational Cost:** Time and memory complexity are high (often \(O(N^{2})\) or \(O(N^{3})\)), making it impractical for very large datasets.

- **Irreversible Decisions:** Agglomerative methods make greedy merges that can't be undone, leading to poor overall clustering if early merges are suboptimal.

- **Sensitivity to Noise & Outliers:** Outliers can significantly influence cluster formation, distorting the hierarchy and leading to misleading results.

- **Difficulty with Large Cluster Numbers:** Interpreting dendrograms becomes challenging when trying to find many clusters, and the structure can be unstable.

- **Choice of Metrics/Linkage:** The final cluster structure heavily depends on the selected distance metric and linkage criteria (e.g., single, complete, average), requiring domain knowledge.

- **No Objective Function:** Unlike some other methods, it lacks a single objective function, relying more on visual interpretation of the dendrogram to decide on the number of clusters.


###8. Why is feature scaling important in clustering algorithms like K-Means?

- **Feature scaling** is crucial in K-Means because it's a distance-based algorithm, and features with larger numerical ranges (like income) will disproportionately influence the distance calculations, overshadowing features with smaller ranges (like age) and leading to biased, inaccurate clusters. Scaling ensures all features contribute equally, improving performance, faster convergence, and preventing the algorithm from being biased by feature magnitude rather than true similarity, making clusters more meaningful.

###Why Scaling Matters in K-Means:

- **Prevents Feature Dominance:** K-Means uses Euclidean distance (based on squared differences) to group points. A feature with a large range (e.g., salary in thousands) will have a much larger impact on the distance than a feature with a small range (e.g., age in tens), even if age is more important for clustering.

- **Ensures Equal Contribution:** Scaling brings all features to a similar scale (e.g., 0 to 1 or a standard normal distribution), so they contribute proportionally to the distance metric, treating all features as equally important.

- **Improves Convergence:** When features are on different scales, the cost function landscape can be elongated, causing the optimization process (like gradient descent used internally) to take longer, zig-zagging to find the minimum. Scaling helps it converge faster to the optimal centroids.

- **Maintains Cluster Quality:** Without scaling, the clusters formed might not accurately reflect the underlying data structure because the distance measure is skewed, leading to poor segmentation.

###Example:

- Imagine clustering people by Age (0-100) and Annual Income (0-1,000,000). A difference of 10 in age is small, but a difference of 10,000 in income is significant. Without scaling, income variations will overwhelmingly dictate which cluster a person belongs to, making age irrelevant to the clustering outcome. Scaling ensures both age and income differences are weighed fairly.


###9. How does DBSCAN identify noise points ?

- **DBSCAN identifies** noise points as those that aren't part of any dense region, specifically points that are neither core points (having enough neighbors within a radius, epsilon) nor border points (reachable from a core point but not dense enough themselves). Essentially, any point left unassigned after forming clusters from dense, core-point-connected areas is labeled as noise (or an outlier) because it lies in low-density space.

###How DBSCAN finds noise points:

1. **Defines Core Points:** The algorithm checks each point. If a point (and its epsilon neighborhood) contains at least min_samples (including itself), it's a core point.

2. **Builds Clusters:** It starts clusters from these core points, adding all density-reachable points (points connected through a chain of core points).

3. **Identifies Border Points:** Points within a core point's epsilon-neighborhood but not dense enough to be core points themselves become border points.

4. **Labels Noise:** Any point that isn't a core point and isn't reachable from any core point (i.e., not a border point) is classified as noise, typically assigned a label of -1.


###Key Parameters: epsilon (Œµ) & min_samples:

- **epsilon (Œµ):** The maximum distance to define a neighborhood.

- **min_samples:** The minimum number of points required in epsilon's neighborhood for a point to be considered a core point.

Points falling outside these dense structures are the outliers, effectively identifying noise during the clustering process itself.


###10. Define inertia in the context of K-Means.

- In the context of **K-Means clustering**, inertia, also known as the Within-Cluster Sum of Squares (WCSS) or the squared error function, is a metric that measures the compactness and internal cohesion of the clusters.

### Definition:

- Inertia is calculated as the sum of the squared Euclidean distances between each data point and the centroid of the cluster to which it has been assigned. The K-Means algorithm works by attempting to minimize this value during the optimization process.

¬†Formally, the inertia for a set of clusters is defined as:


$J$=$\sum _{i=1}^{k}\sum _{x\in C_{i}}||x-\mu _{i}||^{2})$


where:

- $k$ is the number of clusters.

- \$C_{i}$ is  the ith cluster.

- x is a data point in cluster \$C_{i}$.

- $ùúá_{i}$ is the centroid (mean) of cluster $C_{i}$
.

- $||...||^{2}$ is the squared Euclidean distance.

###Interpretation and Use:

A lower inertia value indicates that the data points within each cluster are closer to their respective centroids, suggesting more compact and "better" defined clusters.

Inertia is primarily used in the Elbow Method to help determine the optimal number of clusters (\(k\)) for a dataset.

- The inertia is calculated for a range of \(k\) values.

- A plot of inertia against the number of clusters (\(k\)) typically shows a steep drop in inertia as \(k\) increases, after which the decrease slows down significantly, forming an "elbow" shape.

- The "elbow" point on the graph is often chosen as the optimal \(k\), representing a good balance between a low inertia score and a reasonable number of clusters.

###11.What is the elbow method in K-Means clustering ?

### Elbow Method in K-Means Clustering:

- The **Elbow Method** is a **heuristic technique** used to determine the **optimal number of clusters (K)** in K-Means clustering.


###  Core Idea:

* K-Means minimizes **Within-Cluster Sum of Squares (WCSS)**, also called **inertia**.
* As **K increases**, WCSS **always decreases** because clusters become smaller and tighter.
* The goal is to find a value of **K after which the improvement becomes marginal**.


###  How It Works:

1. Run K-Means for a range of values (e.g., K = 1 to 10).
2. Compute **WCSS** for each K.
3. Plot **K vs WCSS**.
4. Identify the point where the curve shows a **sharp bend (elbow)**.

This ‚Äúelbow‚Äù indicates a **good trade-off** between cluster compactness and model complexity.


###  Why It Helps:

* Avoids choosing too few clusters (underfitting).
* Avoids choosing too many clusters (overfitting).
* Provides a **visual and intuitive** way to select K.


###  Practical Example:

**Customer Segmentation**

* If WCSS drops sharply until **K = 4** and then flattens, **4 clusters** is likely optimal.


###  Limitations

* The elbow is sometimes **not clearly visible**.
* It is **subjective** and depends on visual interpretation.
* Often combined with **Silhouette Score** for better validation.

###12. Describe the concept of "density" in DBSCAN.

- In DBSCAN, density defines clusters as dense regions of points separated by sparser areas, using two parameters: epsilon (Œµ) (neighborhood radius) and MinPts (minimum points in that radius) to classify points as core, border, or noise, enabling it to find arbitrary shapes and handle outliers.

###Key Concepts of Density in DBSCAN:

1. **Epsilon (Œµ) (Œµ-neighborhood)**: This is the maximum distance to consider a point a neighbor. It defines the "reachability" radius.

2. **MinPts (Minimum Points):** The minimum number of points (including the point itself) required within an Œµ-neighborhood for that point to be considered a core point.

3. **Core Point:** A point with at least MinPts neighbors within distance Œµ. These points form the dense heart of a cluster.

4. **Density Reachability:** A point p is density-reachable from a core point q if p is within the Œµ-neighborhood of q, and q is a core point.

5. **Density Connectivity: **A chain of density-reachable points forms a connected component, which becomes a cluster.

6. Border Point: A point that isn't a core point but is within the Œµ-neighborhood of a core point.

7. Noise Point: A point that is neither a core point nor a border point; it lies in low-density regions and is considered an outlier.

###How Density Works:

- DBSCAN scans for core points, then expands clusters by finding density-reachable neighbors, effectively grouping tightly packed areas and isolating sparse points as noise, discovering clusters of complex shapes.

###13. Can hierarchical clustering be used on categorical data ?
- Yes, hierarchical clustering can absolutely be used on categorical data, but it requires using specific distance/dissimilarity measures (like Hamming or Jaccard distance) instead of standard Euclidean distance, or encoding the data appropriately, to measure similarity between categories, enabling algorithms like Agglomerative Clustering to group items based on shared attributes rather than numerical proximity.

###How it Works:

1. **Distance Metric is Key :** Since categorical data (like colors, types) isn't numerical, you can't use standard means; you need a way to quantify how "different" two items are.

    - **Simple Matching Coefficient:** Counts the proportion of attributes where two items have different categories.

    - **Hamming Distance:** Counts the number of positions where categories differ (useful for binary or nominal data).

    - **Jaccard Distance:** Measures dissimilarity based on the absence/presence of shared categories.

2. **Dissimilarity Matrix:** These metrics create a matrix showing the distance between every pair of data points.

3. **Agglomerative Process:** The algorithm starts with each item as its own cluster and iteratively merges the two closest clusters (based on the dissimilarity matrix) until a single cluster remains, forming a hierarchy visualized by a dendrogram.

###15.Explain the term "linkage criteria" in hierarchical clustering.

- In hierarchical clustering, **linkage criteria** define how the distance between two clusters is measured, guiding the algorithm on which clusters to merge (agglomerative) or split (divisive) at each step, fundamentally shaping the final cluster hierarchy (dendrogram). Common methods include Single (nearest points), Complete (farthest points), Average (all pairs average), and Ward's (minimizing variance), each influencing cluster shape and effectiveness based on data characteristics.


###Common Linkage Methods:

- **Single Linkage (Nearest Neighbor)**: Distance is the minimum distance between any point in one cluster and any point in the other cluster. Good for finding elongated clusters but can be sensitive to noise.

- **Complete Linkage (Farthest Neighbor):** Distance is the maximum distance between any point in one cluster and any point in the other. Tends to produce more compact, spherical clusters.

- **Average Linkage (UPGMA):** Distance is the average of all pairwise distances between points from the two clusters. A compromise between single and complete linkage.

- **Ward's Method:** Minimizes the increase in total within-cluster variance (sum of squared errors) when merging two clusters, often leading to clusters of similar size and density.

- **Centroid Linkage:** Distance is the distance between the centroids (mean points) of the two clusters.


###Why It Matters:

- **Cluster Shape:** Different methods create clusters of varying shapes (e.g., elongated vs. compact).

- **Data Sensitivity:** Some methods handle noise or outliers better than others.

- **Interpretation:** Choice affects the resulting hierarchy and how you interpret the dendrogram, which visualizes the clustering process.

Choosing the right linkage method depends on the data's nature and the desired outcome, making it a crucial step in hierarchical clustering.

###16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?

- K-Means clustering performs poorly on data with varying cluster sizes or densities primarily because its underlying objective function and assumptions favor clusters that are roughly spherical, equal in size, and evenly dense.

###Here are the key reasons why this is the case:

- Reliance on Means: K-Means works by minimizing the variance within each cluster (sum of squared distances from points to the cluster mean). This process inherently defines the "center" of a cluster as the arithmetic mean. In clusters with vastly different densities or shapes, the mean might not accurately represent the true center or the natural grouping of the data.

- Assumption of Spherical Clusters: The algorithm implicitly assumes that clusters are convex and spherical. This means it struggles to identify clusters with non-spherical shapes (e.g., elongated or crescent-shaped), which are common in real-world data with varying densities and sizes.

- Bias Towards Larger/Denser Clusters: The minimization of within-cluster variance means the algorithm is sensitive to the overall distribution of points. Denser or larger clusters can exert a stronger influence on the location of cluster centers (centroids), potentially causing the algorithm to "steal" points from smaller or less dense clusters to reduce the overall error.

- Centroid Displacement: With differing densities, the calculated centroid of a sparse cluster may be pulled toward the proximity of a dense cluster, leading to misclassification of boundary points.

- Equal Importance Assumption: K-Means treats all dimensions and all clusters as having equal importance and structure. This lack of flexibility prevents it from adapting to the local characteristics of individual clusters, such as their unique size or density.


In summary, when data deviates from the ideal scenario of uniform, spherical, and equally-sized clusters, K-Means tends to misrepresent the natural data structure, leading to suboptimal or incorrect clustering results. For such datasets, alternative algorithms like DBSCAN (Density-Based Spatial Clustering of Applications with Noise) or Gaussian Mixture Models (GMM) are often more appropriate, as they can adapt to different cluster shapes, sizes, and densities .


###17. What are the core parameters in DBSCAN, and how do they influence clustering ?

- **DBSCAN's** core parameters are eps (epsilon), defining neighborhood radius, and minPts (minimum points), the minimum density to form a cluster; eps controls cluster spread (smaller splits, larger merges), while minPts dictates density (higher makes it stricter, lower finds smaller groups), both crucial for identifying dense regions (core points) vs. noise/outliers and shaping the final cluster boundaries.

###Core Parameters:

1. eps (Epsilon - Œµ):

    - Definition: The maximum distance between two points for one to be considered a neighbor of the other.

    - Influence:
                    
      - Small eps: Can split one dense cluster into several, or classify many points as noise.

      - Large eps: Can merge distinct clusters into one large one, losing granularity.

2. minPts (Minimum Points):

    - Definition: The minimum number of points required within the eps-neighborhood for a point to be considered a "core point" (a point in a dense region).

    - Influence:

      - High minPts: Makes the algorithm stricter, requiring more points for a dense region, potentially leading to fewer but more robust clusters and more noise.

      - Low minPts: Can result in smaller clusters and might misclassify noise as part of clusters, as the density requirement is loose.


###How They Work Together:

- Core Point: A point with at least minPts neighbors (including itself) within distance eps.

- Border Point: A point within a cluster's eps-neighborhood but not a core point (fewer than minPts neighbors).

- Noise Point: A point that isn't a core point and isn't reachable from any core point.

By tuning eps and minPts, you control how "dense" a region must be to be considered a cluster and how far points can be to still belong, effectively managing cluster size, shape, and noise detection.


###18. How does K-Means++ improve upon standard K-Means initialization ?

- K-Means++ improves standard K-Means by choosing initial centroids intelligently, selecting the first one randomly and subsequent ones with a probability proportional to their squared distance from existing centers, ensuring better spread and faster convergence to a high-quality solution, unlike standard K-Means which randomly places all centers, often leading to poor, inconsistent results.

### Standard K-Means Initialization:

- **Random Placement:** All \(K\) initial centroids are chosen randomly from the data points.

- **Drawback:** This can result in centroids being too close, overlapping, or even leading to empty clusters, trapping the algorithm in a poor local minimum.

###K-Means++ Initialization:

1. **First Centroid:** Select the first centroid uniformly at random from the dataset.

2. **Subsequent Centroids (D¬≤ Sampling):** For each remaining centroid :

    1. Calculate the squared distance (D¬≤) of every data point to its nearest existing centroid.
    
    2. Select the next centroid from the data points, with a probability proportional to this D¬≤ value.
    
3. **Result :** This "farthest-first" probabilistic approach ensures initial centers are well-spaced, reducing overlap and making clusters more distinct from the start.

###Key Improvements:

- **Better Quality:** Leads to more stable, accurate, and consistent clustering results.

- **Faster Convergence:** By starting with a good spread, the algorithm typically reaches the optimal solution in fewer iterations.

- **Robustness:** Handles varying cluster densities and shapes better than random initialization.


###19. What is agglomerative clustering ?

- **Agglomerative clustering** is a "bottom-up" hierarchical clustering method that starts with each data point as its own cluster and iteratively merges the closest pairs of clusters until all points belong to a single, large cluster, creating a hierarchy visualized by a dendrogram (tree diagram). It uses distance metrics (like Euclidean distance) and linkage criteria (single, complete, average) to decide which clusters to merge, allowing for interpretation of data structure without needing to pre-specify the number of clusters.

###How it works :

1. **Initialization :** Treat every single data point as its own cluster.

2. **Distance Calculation :** Compute the distance (or similarity) between all pairs of clusters (initially, between all data points) using a chosen metric (e.g., Euclidean, Manhattan).

3. **Merging :** Merge the two closest clusters based on a specific linkage rule (e.g., single linkage: shortest distance; complete linkage: longest distance; average linkage: average distance).

4. **Update :** Recalculate distances involving the newly formed cluster.

5. **Repeat :** Continue merging until only one cluster remains, containing all data points.


###Key Characteristics & Benefits :

- **Hierarchical Structure:** Produces a tree-like hierarchy (dendrogram) showing how clusters form at different levels of similarity.

- **No Predefined 'k':** Unlike K-Means, you don't need to specify the number of clusters (k) beforehand; the dendrogram helps determine the best number.

- **Interpretability:** The dendrogram offers a clear visual representation of relationships between data points.

###When to Use It :

- When you want to understand the nested structure of your data.

- For smaller datasets where computational cost isn't a major issue.

- In domains like image segmentation, customer segmentation, or biological data analysis.


###20. What makes Silhouette Score a better metric than just inertia for model evaluation ?

- **Silhouette Score** is better than just Inertia (WCSS) because it measures both cluster cohesion (compactness) and separation, providing a holistic view, whereas Inertia only assesses internal cluster compactness, potentially leading you to choose too many clusters if used alone (as seen in the Elbow Method). A high Silhouette Score (closer to 1) signifies well-defined, distinct clusters, while Inertia always decreases with more clusters, offering no clear optimal point without visual interpretation.

###Key Differences & Advantages of Silhouette Score:

- Cohesion & Separation:

    - Inertia (WCSS): Measures the sum of squared distances of samples to their closest cluster center (compactness). It doesn't consider how far clusters are from each other.

    - Silhouette Score: Calculates how similar a point is to its own cluster (cohesion) compared to other clusters (separation). It combines both aspects into one score.

- Interpretation:

    - Inertia: Lower is better, but it keeps dropping as you add more clusters (K), making it hard to find the "elbow" (e.g., in the Elbow Method).

    - Silhouette Score: Ranges from -1 to +1. A score near +1 means good clusters, near 0 means points are on boundaries, and negative means misclassified points.

- Model Comparison :
    - Silhouette Score: Excellent for comparing different clustering models (e.g., different K values in K-Means) because it gives a direct measure of overall structure quality.

- Problem Identification:

    - A negative Silhouette Score for some points indicates they might be better off in a different cluster, a flaw Inertia doesn't reveal.

###When to Use Which?

- Use Inertia (Elbow Method): To find an initial, rough estimate of the optimal number of clusters by observing where the rate of inertia decrease slows down.


- Use Silhouette Score: To validate and refine the cluster count by selecting the K that yields the highest average score, confirming good separation and compactness.

In essence, Inertia tells you how tight your clusters are, while Silhouette Score tells you how well-defined and separated those tight clusters are, making it a more robust metric for overall clustering quality.

###29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

###31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate overlapping data
X, _ = make_blobs(
    n_samples=400,
    centers=3,
    cluster_std=2.0,
    random_state=42
)

# Fit K-Means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Create a meshgrid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

# Predict cluster for each point in the meshgrid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries and data points
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=200)
plt.title("K-Means Clustering with Decision Boundaries (Overlapping Clusters)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

###34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# 2. Initialize a StandardScaler object
scaler = StandardScaler()

# 3. Fit the scaler to X and transform the data
X_scaled = scaler.fit_transform(X)


# Initialize an empty list to store inertia values
inertia_values = []

# Loop through K values from 2 to 6
for k in range(2, 7):
    # Initialize KMeans with the current k, random_state, and n_init
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)

    # Fit K-Means to the scaled data
    kmeans.fit(X_scaled)

    # Append the inertia_ (WCSS) to the list
    inertia_values.append(kmeans.inertia_)

# Print the collected inertia values
print("Inertia values for K=2 to 6:", inertia_values)



# Create a list of K values (from 2 to 6) corresponding to the inertia_values
k_values = list(range(2, 7))

# Generate the line plot
plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia_values, marker='o', linestyle='-')
plt.title('K-Means Inertia (WCSS) vs. Number of Clusters (K)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.xticks(k_values) # Ensure x-axis ticks are only at integer K values
plt.show()

print("Line plot of Inertia vs. K generated successfully.")
```

###44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.


```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler

# Load Iris dataset
iris = load_iris()
X = iris.data

# Standardize features
X_scaled = StandardScaler().fit_transform(X)

# Apply KMeans
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Compute silhouette scores
silhouette_avg = silhouette_score(X_scaled, cluster_labels)
sample_silhouette_values = silhouette_samples(X_scaled, cluster_labels)

# Plot silhouette coefficients
plt.figure(figsize=(8, 6))
y_lower = 10

for i in range(n_clusters):
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_silhouette_values
    )

    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, f"Cluster {i}")
    y_lower = y_upper + 10

# Average silhouette score line
plt.axvline(x=silhouette_avg, linestyle="--")
plt.title("Silhouette Plot for KMeans Clustering (Iris Dataset)")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Samples")
plt.show()
```

###47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate blob data
X, _ = make_blobs(
    n_samples=600,
    centers=4,
    cluster_std=1.2,
    random_state=42
)

# Add uniform noise
rng = np.random.RandomState(42)
noise = rng.uniform(low=-10, high=10, size=(80, 2))
X_noisy = np.vstack([X, noise])

# Standardize features
X_scaled = StandardScaler().fit_transform(X_noisy)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.35, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters and noise points
n_noise = np.sum(labels == -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {n_noise}")

# Visualization
plt.figure(figsize=(7, 6))
plt.scatter(
    X_scaled[:, 0],
    X_scaled[:, 1],
    c=labels,
    s=30
)
plt.title("DBSCAN Clustering with Noise Points")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

###48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

# Load Digits dataset
digits = load_digits()
X = digits.data
y = digits.target  # true labels (not used for clustering)

# Standardize features (important before t-SNE)
X_scaled = StandardScaler().fit_transform(X)

# t-SNE for dimensionality reduction
tsne = TSNE(
    n_components=2,
    perplexity=30,
    learning_rate=200,
    n_iter=1000,
    random_state=42
)
X_tsne = tsne.fit_transform(X_scaled)

# Agglomerative Clustering
agglo = AgglomerativeClustering(
    n_clusters=10,   # digits 0‚Äì9
    linkage='ward'
)
cluster_labels = agglo.fit_predict(X_tsne)

# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(
    X_tsne[:, 0],
    X_tsne[:, 1],
    c=cluster_labels,
    s=15
)
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE 2D)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(label="Cluster")
plt.grid(True)
plt.show()
```
